```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"
df = pd.read_csv(filepath)
```

handling missing values

```python
df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']],2)
df.head()
```

```python
df= df.replace('?',np.NaN)
```

```python
missing_values=df.isnull()
print(missing_values.head())
```

```python
for column in missing_values.columns.values.tolist():
    print(column)
    print (missing_values[column].value_counts())
    print("")
```

```python
# wieght_kg, ScreenSize

# fixing missing values

wieght_mean= df['Weight_kg'].mean(axis=0)
df['Weight_kg'].replace(np.NaN,wieght_mean,inplace=True)
```

```python
freq1=df['Screen_Size_cm'].mode()
df['Screen_Size_cm'].replace(np.NaN,float(freq1),inplace=True)
```

```python
df.head(50)
```

```python
df.dtypes
```

Data Standardization


```python
# cm to inches
df['Screen_Size_cm']= df['Screen_Size_cm']/2.54
df.head(5)
```

```python
# kg to pounds
df['Weight_kg']=df['Weight_kg']*2.205
df.head()
```

```python
df.rename(columns={'Weight_kg':'Weight_pound'},inplace=True)
```

```python
df.rename(columns={'Screen_Size_cm':'Screen_size_inches'},inplace=True)
```

```python
df.columns
```

Data Normalization

```python
df['CPU_frequency']=df['CPU_frequency']/df['CPU_frequency'].max()
```

```python
df.head()
```

Data Binning

```python
labless=[ "Low", "Medium" , "High"]
```

```python
bins= np.linspace(min(df['Price']),max(df['Price']),4,)
df['Price_category']=pd.cut(df['Price'],bins,labels=labless,include_lowest=True)
```

```python

plt.bar(labless, df["Price_category"].value_counts())
plt.xlabel("Price")
plt.ylabel("count")
plt.title("Price bins")

```

```python
dummy1=pd.get_dummies(df['Screen'])
dummy1.rename(columns={'Full HD':'IsIPL','IPS Panel':'IsLCD'},inplace=True)

```

```python
df=pd.concat([df,dummy1],axis=1)
df.head()
```

```python
df.drop("Screen", axis = 1, inplace=True)
```

```python
df.drop("Unnamed: 0", axis = 1, inplace=True)
```
