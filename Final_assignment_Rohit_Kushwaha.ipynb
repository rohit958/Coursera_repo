```python
# importing libaries
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
# graphh funtion for question 5 & 6

def make_graph(stock_data, revenue_data, stock):

    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .3)

    stock_data_specific = stock_data[stock_data.Date <= '2021--06-14']

    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']

    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date, infer_datetime_format=True), y=stock_data_specific.Close.astype("float"), name="Share Price"), row=1, col=1)

    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Date, infer_datetime_format=True), y=revenue_data_specific.Revenue.astype("float"), name="Revenue"), row=2, col=1)

    fig.update_xaxes(title_text="Date", row=1, col=1)

    fig.update_xaxes(title_text="Date", row=2, col=1)

    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)

    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)

    fig.update_layout(showlegend=False,

    height=900,

    title=stock,

    xaxis_rangeslider_visible=True)

    fig.show()
```

#### Question 1 - Extracting Tesla Stock Data Using yfinance

```python
#connection to yahoo finace using Ticker
Tesla_html=yf.Ticker('TSLA')
tesla_data=Tesla_html.history(period="max")
tesla_data.reset_index(inplace=True)

#first five rows
tesla_data.head()
```

#### Question 2 - Extracting Tesla Revenue Data Using Webscraping

```python
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
```

```python
#loading data using website
tesla_html=requests.get(url).text

#creating soup
tesla_soup=BeautifulSoup(tesla_html,'html5lib')

#creating Dataframe
cols=['Date','Revenue']
tesla_revenue=pd.DataFrame(columns=cols)

#loading soup data into dataframe
for row in tesla_soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    Date = col[0].text
    Revenue = col[1].text

    tesla_revenue=tesla_revenue._append({'Date':Date,'Revenue':Revenue}, ignore_index=True)

#first five rows
tesla_revenue.head()
```

#### Question 3 - Extracting GameStop Stock Data Using yfinance

```python
#loading data using
game=yf.Ticker('GME')

#loading data into dataframe
gme_data=game.history(period='max')
gme_data.reset_index(inplace=True)

#first five rows
gme_data.head()
```

#### Question 4 - Extracting GameStop Revenue Data Using Webscraping

```python
#loading html data from website
gmeurl='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html'
gme=requests.get(gmeurl).text

#creating soup
gme_soup=BeautifulSoup(gme,'html5lib')

#creating dataframe
cols=['Date','Revenue']
gme_revenue=pd.DataFrame(columns=cols)

#loading data into dataframe
for row in gme_soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    Date = col[0].text
    Revenue = col[1].text

    gme_revenue=gme_revenue._append({'Date':Date,'Revenue':Revenue}, ignore_index=True)

#first 5 rows
gme_revenue.head()
```

#### Question 5 - Tesla Stock and Revenue Dashboard

```python
#cleaning data
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace('$',"").str.replace(',',"")
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]

#calling funtion define to plot graph
make_graph(tesla_data,tesla_revenue,'Tesla')
```

#### Question 6 - GameStop Stock and Revenue Dashboard

```python
#cleaning data
gme_revenue["Revenue"] = gme_revenue['Revenue'].str.replace('$',"").str.replace(',',"")
gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]

#calling funtion define to plot graph
make_graph(gme_data,gme_revenue,'Gamestop')
```
